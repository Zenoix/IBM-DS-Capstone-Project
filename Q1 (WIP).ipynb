## Import libraries and dataset

```python
import pandas as pd

NY331_df = pd.read_csv("311_Service_Requests_from_2010_to_Present_min.csv")
NY331_df.drop("Unnamed: 0", axis=1, inplace=True)
```

## Briefly explore and clean dataset

```python
NY331_df.head()
```

```python
# Check number of rows (number of complaints) and columns
print(NY331_df.shape)
print("Number of complaints:", NY331_df.shape[0])
```

```python
# make sure the datatypes seem reasonable
NY331_df.info()
```

```python
# Convert the created and closed date into datetime type
NY331_df["Created Date"] = pd.to_datetime(NY331_df["Created Date"], format="%m/%d/%Y %I:%M:%S %p")
NY331_df["Closed Date"] = pd.to_datetime(NY331_df["Closed Date"], format="%m/%d/%Y %I:%M:%S %p")
NY331_df.info()
```

```python
# Check how many different complaint types there are and if there are duplicates
print(NY331_df["Complaint Type"].value_counts())
```

```python
# Using 800,000 as the threshold, what complaint type(s) do you recommend the Department of Housing Preservation and Development of New York City address first?\
NY331_df["Complaint Type"].value_counts() >= 800000
```
